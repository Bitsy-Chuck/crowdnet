```python
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.models import HoverTool
from bokeh.palettes import Dark2
import numpy as np
import os

palette = Dark2[3]

hover = HoverTool(
        tooltips="""
        <div>
            <div>
                <img
                    src="@images" height="120" alt="@images" width="176"
                    style="float: left; margin: 0px 15px 15px 0px;"
                    border="2"
                ></img>
                <img
                    src="@predicted_images" height="120" alt="@predicted_images" width="176"
                    style="float: left; margin: 0px 15px 15px 0px;"
                    border="2"
                ></img>
                <img
                    src="@true_images" height="120" alt="@true_images" width="176"
                    style="float: left; margin: 0px 15px 15px 0px;"
                    border="2"
                ></img>
            </div>
            <div>
                <span style="font-size: 15px; color: #966;">[$index]</span>
            </div>
            <div>
                <span style="font-size: 15px;">Location</span>
                <span style="font-size: 10px; color: #696;">($x, $y)</span>
            </div>
        </div>
        """
    )

predicted_person_count = np.load('visualization/train/predicted_person_count.npy')
true_person_count = np.load('visualization/train/true_person_count.npy')
image_directory = 'visualization/train/original_images'
image_paths = ['{}/{}'.format(image_directory, file) for file in os.listdir(image_directory) if file.endswith('.jpeg')]
predicted_directory = 'visualization/train/predicted_label_images'
predicted_paths = ['{}/{}'.format(predicted_directory, file) for file in os.listdir(predicted_directory) if file.endswith('.jpeg')]
true_directory = 'visualization/train/true_label_images'
true_paths = ['{}/{}'.format(true_directory, file) for file in os.listdir(true_directory) if file.endswith('.jpeg')]

train_source = ColumnDataSource(
        data=dict(
            x=true_person_count,
            y=predicted_person_count,
            images=image_paths,
            predicted_images=predicted_paths,
            true_images=true_paths
        )
    )

predicted_person_count = np.load('visualization/test/predicted_person_count.npy')
true_person_count = np.load('visualization/test/true_person_count.npy')
image_directory = 'visualization/test/original_images'
image_paths = ['{}/{}'.format(image_directory, file) for file in os.listdir(image_directory) if file.endswith('.jpeg')]
predicted_directory = 'visualization/test/predicted_label_images'
predicted_paths = ['{}/{}'.format(predicted_directory, file) for file in os.listdir(predicted_directory) if file.endswith('.jpeg')]
true_directory = 'visualization/test/true_label_images'
true_paths = ['{}/{}'.format(true_directory, file) for file in os.listdir(true_directory) if file.endswith('.jpeg')]

test_source = ColumnDataSource(
        data=dict(
            x=true_person_count,
            y=predicted_person_count,
            images=image_paths,
            predicted_images=predicted_paths,
            true_images=true_paths
        )
    )


plot = figure(plot_width=800, plot_height=800, tools="pan,wheel_zoom,box_zoom,reset,save",
              title="Person Count", x_axis_label='Predicted', y_axis_label='True')
plot.add_tools(hover)
plot.circle('x', 'y', size=10, source=train_source, fill_color=palette[0], line_color=palette[0])
plot.square('x', 'y', size=15, source=test_source, fill_color=palette[1], line_color=palette[1])
show(plot)
```
