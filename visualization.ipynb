```python
from bokeh.plotting import figure, output_file, show, ColumnDataSource
from bokeh.models import HoverTool, BoxZoomTool, ResetTool
import numpy as np
import os

predicted_person_count = np.load('visualization/predicted_person_count.npy')
true_person_count = np.load('visualization/true_person_count.npy')
image_directory = 'visualization/original_images'
image_paths = ['{}/{}'.format(image_directory, file) for file in os.listdir(image_directory) if file.endswith('.jpeg')]
predicted_directory = 'visualization/predicted_label_images'
predicted_paths = ['{}/{}'.format(predicted_directory, file) for file in os.listdir(predicted_directory) if file.endswith('.jpeg')]
true_directory = 'visualization/true_label_images'
true_paths = ['{}/{}'.format(true_directory, file) for file in os.listdir(true_directory) if file.endswith('.jpeg')]

source = ColumnDataSource(
        data=dict(
            x=true_person_count,
            y=predicted_person_count,
            images=image_paths,
            predicted_images=predicted_paths,
            true_images=true_paths
        )
    )

hover = HoverTool(
        tooltips="""
        <div>
            <div>
                <img
                    src="@images" height="120" alt="@images" width="176"
                    style="float: left; margin: 0px 15px 15px 0px;"
                    border="2"
                ></img>
                <img
                    src="@predicted_images" height="120" alt="@predicted_images" width="176"
                    style="float: left; margin: 0px 15px 15px 0px;"
                    border="2"
                ></img>
                <img
                    src="@true_images" height="120" alt="@true_images" width="176"
                    style="float: left; margin: 0px 15px 15px 0px;"
                    border="2"
                ></img>
            </div>
            <div>
                <span style="font-size: 15px; color: #966;">[$index]</span>
            </div>
            <div>
                <span style="font-size: 15px;">Location</span>
                <span style="font-size: 10px; color: #696;">($x, $y)</span>
            </div>
        </div>
        """
    )

plot = figure(plot_width=800, plot_height=800, tools="pan,wheel_zoom,box_zoom,reset,save",
              title="Person Count", x_axis_label='Predicted', y_axis_label='True')
plot.add_tools(hover)
plot.circle('x', 'y', size=15, source=source)
show(plot)
```
